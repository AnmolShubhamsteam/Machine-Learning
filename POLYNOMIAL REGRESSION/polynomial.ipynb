```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures as PF
```

```python
data=pd.read_csv("Position_Salaries (1).csv")
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
x=np.array(data.iloc[:,1]).reshape(-1,1)
x
```

```python
y=np.array(data.iloc[:,-1]).reshape(-1,1)
y
```

```python
from sklearn.linear_model import LinearRegression as LR
```

```python
lr=LR()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,"*")
plt.plot(x,lr.predict(x))
```

```python
lr.intercept_
```

```python
lr.predict([[6.5]])
```

```python
poly_regressor=PF(degree=5)
```

```python
lr.predict(x)
```

```python
x_poly=poly_regressor.fit_transform(x)
x_poly
```

```python
poly_regressor_main=LR()
```

```python
poly_regressor_main.fit(x_poly,y)
```

```python
plt.plot(x,y,"*")
plt.plot(x,poly_regressor_main.predict(x_poly))
```
